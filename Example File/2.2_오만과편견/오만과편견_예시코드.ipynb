###데이터 가공하기

```python
c1 = char1[['sociable','talkative','dancing']].to_numpy()
print(c1)
```

```python
c1_x = np.column_stack((c1[:,0], np.mean(c1, axis=1)))
print(c1_x)
```

```python
c2 = char2[['detailed','realistic','thorough']].to_numpy()
c3 = char3[['cooperative','sensitive','expressive']].to_numpy()
c4 = char4[['stable','planned','diligent']].to_numpy()

c2_x = np.column_stack((c2[:,0], np.mean(c2, axis=1)))
c3_x = np.column_stack((c3[:,0], np.mean(c3, axis=1)))
c4_x = np.column_stack((c4[:,0], np.mean(c4, axis=1)))

print(c2_x)
print(c3_x)
print(c4_x)
```

###4. Make! 소설 속 등장 인물의 성격 데이터로 인공지능 모델 만들기

```python
from sklearn.cluster import KMeans

km1 = KMeans(n_clusters = 2, random_state=42)
km1.fit(c1_x)
```

```python
print(km1.labels_)
print(np.unique(km1.labels_, return_counts=True))
```

```python
c1_x[km1.labels_== 0]
```

```python
c1_x[km1.labels_== 1]
```

```python
km2 = KMeans(n_clusters = 2, random_state=42)
km3 = KMeans(n_clusters = 2, random_state=42)
km4 = KMeans(n_clusters = 2, random_state=42)

km2.fit(c2_x)
km3.fit(c3_x)
km4.fit(c4_x)
```

##5. Predict! K-Means 알고리즘을 이용하여 Elizabeth와 Darcy의 MBTI 예측하기

```python
print(km1.cluster_centers_)
```

```python
print(km1.transform(c1_x[0:1]))
print(km1.transform(c1_x[1:2]))
```

```python
print(km1.predict(c1_x[0:1]))
print(km1.predict(c1_x[1:2]))
```

```python
plt.scatter(c1_x[:,0], c1_x[:,1], c=km1.predict(c1_x), cmap='brg' ) #c1_x의 샘플들을 predict()로 0,1로 군집화 시켜 'brg'색의 형태로 분포도를 그림.
center = km1.cluster_centers_ #군집(Cluster)의 중심점을 center에 저장
plt.title('Centroid of c1_x')
plt.scatter(center[:,0],center[:,1], color = 'black', s=100, label = 'Centroid',alpha=0.7) #중심점을 크기 100의 검은색 점으로 그림. alpha는 투명도
plt.scatter(c1_x[0,0], c1_x[0,1], color = 'r', marker = '^', label = 'Elizabeth')
plt.scatter(c1_x[1,0], c1_x[1,1], color = 'b', marker= 'D', label = 'Darcy')
plt.legend()
plt.show()
```

##4. More & More! 최적의 K 찾기(Elbow Method)

```python
plt.scatter(c1_x[:,0], c1_x[:,1]) #c1_x 데이터 좌표 그리기

n_clusters = range(1,10) #n_clusters를 1부터 10까지 1씩 증가시킨다.
kmeans = [KMeans(n_clusters=i) for i in n_clusters]

#모든 샘플의 제곱오차합계(inertia_)를 계산하여 in_1 리스트에 추가한다.
in_1 = [kmeans[i].fit(c1_x).inertia_ for i in range(len(kmeans))] 

plt.plot(n_clusters, in_1)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve1')
plt.show()
```
